### Read Data

```python
movie_ratings = pd.read_csv('../input/movies_rating.csv', header=0, index_col='Movie')
```

```python
#Print the DataFrame just created
print(movie_ratings)
```

### Calculating the cosine similarity across items

#### As the cosine_similarity function returns an array, we create a new data frame and assign the names to columns and rows which come from the original dataframe

```python
name_movies= list(movie_ratings.index)

sim = pd.DataFrame(cosine_similarity(movie_ratings), columns=name_movies, index=name_movies)

print("Similarity Matrx: \n")

print(sim)
```

### Task 1

#### The Top5 similar movies to 'Toy Story (1995)' the highest the correlation value, the more similar. Observe that the dataframe has been sliced from the index 1, since in the index 0 the value will be 1.00 (self-correlation).

```python
print("\nTop 5 similar movies to 'Toy Story (1995)':\n")

corr_top5=pd.DataFrame(sim.loc['Toy Story (1995)'].sort_values(ascending=False)[1:6])

print(corr_top5)
```
